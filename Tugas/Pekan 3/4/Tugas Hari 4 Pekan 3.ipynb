````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
````



---



---



<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk Global_Sales berdasarkan kategori Genre.

````python
#code here

# df_genre = pd.pivot_table(data=df, index='Genre', values='Global_Sales')

df_genre = df.groupby('Genre').mean()

df_genre = df_genre.loc[:,['Global_Sales']]

df_genre
````

<h2>Soal 2: Perbandingan Kuantitatif Pieplot - Part a whole using Pie plot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```
2. Insight yang didapat berdasarkan pie chart berikut yaitu terpapar persentase dari masing-masing Genre dimana persentase terbesar ada pada genre Platform dengan 14.8% dan yang terkecil ada pada genre Adventure dengan 2.9%
```

````python
#code here
plt.pie(df_genre['Global_Sales'], labels=df_genre.index, autopct='%1.1f%%')

plt.show()
````

<h2>Soal 3: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

````python
#code here

df_genre_new = df.groupby('Genre').mean()

df_genre_new = df_genre_new.loc[:,['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]

df_genre_new
````

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```
2. Didapat Mean Sales tertinggi ada pada NA Sales dengan genre Platform dan yang terendah ada pada Other Sales dengn genre Adventure

```

````python
df_genre_new.T.plot(kind='bar', figsize=(18,9))

plt.title("Mean Sales Video Games By Genre")
plt.xlabel("Region Sales")
plt.ylabel("Mean Sales")

plt.xticks(rotation=0)

plt.show()
````
