<h2>Soal 4: Comparison Operator dengan fungsi</h2>

Buatlah sebuah fungsi yang menerima satu argument bertipe data numeric dan menghasilkan sebuah return sebagai berikut :
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan ganjil
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 2 sampai 5 (2 dan 5 termasuk)
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 6 sampai 20 (6 dan 20 termasuk)
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan lebih besari dari 20

```python
# Buat fungsi

def isAneh(angka):
  if angka%2 != 0:
    return 'Aneh'
  elif angka%2 == 0 and (angka>=2 and angka<=5):
    return 'Tidak Aneh'
  elif angka%2 == 0 and (angka>=6 and angka<=20):
    return 'Aneh'
  elif angka%2 == 0 and (angka>20):
    return 'Tidak Aneh'

print(isAneh(22))
```

<h2>Soal 5: While Loop dan For Loop</h2>

- Apa perbedaan while loop dan for loop?
- Berikan contoh sederhana cara menggunakan while loop dan for loop

jawaban perbedaan:

`For Loop`

- Untuk mengeksekusi blok kode tertentu berulang kali sampai beberapa kondisi terpenuhi

- Ketika Anda mengetahui jumlah iterasi sebelumnya

- Ketika Anda memiliki gagasan tentang kisaran nilai yang akan diulang tetapi tidak yakin angka pastinya

`While Loop`

- Akan berulang sampai jumlah iterasi yang disebutkan selesai

- Akan berulang sampai kondisi yang disajikan selesai

```python
# jawaban contoh

x = 5
while x > 0 :
    print(x)
    x-=1

print()

for i in range(5):
  print(i)
```

<h2>Soal 6: While Loop</h2>
    
Buatlah sebuah code while loop sebagai berikut :
- Buatlah suatu fungsi yang menerima satu input nilai numerik
- Dalam fungsi, Buat sebuah while-loop dengan melakukan comparasi terhadap inputan tersebut, dimana looping/iterasi terus berjalan apabila nilai inputan tidak sama dengan 0
- Dalam setiap iterasi update nilai variable inputan itu, jika input lebih dari 0, maka kurangi variabel tersebut sebesar 1, selain itu tambahkan 1
- print nilai variable input dalam setiap awal iterasi

```python
a = 10

def fungsi_while(angka):
  
  if angka > 0:
    while angka != 0:
      print(angka)
      angka-=1
  elif angka < 0:
    while angka != 0:
      print(angka)
      angka+=1

fungsi_while(a)

print()

fungsi_while(-a)
```

<h2>Soal 7: For Loop</h2>

Lengkapi kode untuk menghasilkan suatu output yang diharapkan:





*   Cari siswa mana saja yang memiliki nilai lebih dari sama dengan 80. Masukkan kedalam sebuah list. print hasilnya






```python
data_siswa = [
    {
        "nama":"Budi",
        "nilai": 90
    },
    {
        "nama":"Nina",
        "nilai": 78
    },
    {
        "nama":"Rudi",
        "nilai": 91
    },
    {
        "nama":"Olivia",
        "nilai": 76
    },
    {
        "nama":"Leo",
        "nilai": 80
    },
    {
        "nama":"Liam",
        "nilai": 67
    },
    {
        "nama":"Sheila",
        "nilai": 76
    }
]
```

```python
#code here

new_data_siswa = []
for i in data_siswa:
  if i['nilai'] >= 80:
    new_data_siswa.append(i['nama'])

print(new_data_siswa)
```

Expected output:

['Budi', 'Rudi', 'Leo']
